```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("Sohazur loves the CIS program at Zayed University.")
```

```python
classifier(
    ["Sohazur comes from Bangladesh.", "UAE is a new experience for Sohazur"]
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "Sohazur really prepares for the Deep learning course at Zayed University.",
    candidate_labels=["education", "politics", "business"],
)
```

```python
from transformers import pipeline

generator = pipeline("text-generation")
generator("Sohazur we late for the class on wednesday")
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "In this course, Sohazur loves to participtate but",
    max_length=30,
    num_return_sequences=2,
)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("This course will teach him all about <mask> models.", top_k=2)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Sohazur and I work at Zayed Unieversity in UAe.")
```

```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I Study?",
    context="My name is Sohazur and I study at Zayed University in UAE",
)
```
